## Generando archivos PDF del presupuesto
* Crear el archivo PDF
* Definir un tipo de letra
* Llenamos el documento
* Guardar el entregable 

```python
!pip install fpdf
```

```python
from fpdf import FPDF
```

```python
pdf = FPDF()

pdf.add_page()
pdf.set_font("Arial")

pdf.image("william.png", x=0, y=0, w=210, h=297)


pdf.text(115, 145, "El mejor proyecto con python")
pdf.text(115, 160, horas_estimadas)
pdf.text(115, 175, valor_hora )
pdf.text(115, 190, plazo_estimado )
pdf.text(115, 205, str(valor_total))


pdf.output("Presupuesto.pdf")
```
